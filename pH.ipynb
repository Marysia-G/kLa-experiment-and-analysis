```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data_pH = pd.DataFrame({'vessel': ["O_96_0.25",	"O_96_0.55",	"O_48_0.29",	"O_48_0.56",	"O_24_0.29",	"O_24_0.56",	"□_24_0.55",	"□_24_1.10",	"T_0.25",	"T_0.56",	"F_0.25",	"F_0.56",	"F_1.12"],
                        'pH': [5.2,	4.5,	4.5,	4.5,	3.6,	4,	3.6,	6.8,	5.2,	4.5,	4.5,	4,	3.6]})

print(data_pH)
```

```python
sns.set_theme(style='white', context='talk')

plt.figure(figsize=(7,7))
sns.barplot(data_pH, x = data_pH['vessel'], y = data_pH['pH'], hue = data_pH['pH'], palette='flare', legend=False)
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('pH')
plt.title('pH')
plt.tight_layout()

plt.savefig('pH')
```
